```python
import pandas as pd

# Load transaction data
transactions = pd.read_csv("data/train_transaction.csv")

# Load identity data
identity = pd.read_csv("data/train_identity.csv")

# Merge both datasets on 'TransactionID'
df = transactions.merge(identity, on="TransactionID", how="left")

# Display basic info
print("Dataset Preview:")
print(df.head())  # Show first few rows
print("\nDataset Info:")
print(df.info())  # Check column types and missing values
print("\nSummary Statistics:")
print(df.describe())  # Summary statistics
```

```python
# saved mereged datasets
df.to_csv("data/merged_fraud_data.csv", index=False)
```
